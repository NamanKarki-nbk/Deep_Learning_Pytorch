```python
import spacy
```

```python
corpus = """Alice went to the Wonderland to explore the mysterious garden. 
She saw strange creatures and colorful flowers. 
"Hello," said a white rabbit, "come with me!" 
Visit https://example.com/wonderland and www.socialmedia.com/alice for more details.
"""
```

```python
#word level tokenization
nlp = spacy.load("en_core_web_sm")
doc = nlp(corpus)
word_tokens = [token.text for token in doc]
print(word_tokens)
```

```python
#character level tokenization
char_tokens = list(corpus)
print(char_tokens)
```

```python
#stopwords removal

filtered_tokens = [token.text for token in doc if not token.is_stop and token.is_alpha]
print(filtered_tokens)
```

```python
#lemmatization
root_words = [token.lemma_ for token in doc]
print(root_words)
```
