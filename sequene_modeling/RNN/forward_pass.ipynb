This is the exmaple of forward pass of RNN using numpy build from scratch.

```python
import numpy as np
```

```python
#input sequence of length 5
x1 = np.random.rand(5)
X2 = np.random.rand(5)
x3 = np.random.rand(5)

print(x1)
print(X2)
print(x3)
```

```python
#hidden size of 4
h0 = np.random.rand(4)

#initialize weights 
Wxh = np.random.rand(4,5)#input to hidden
Whh = np.random.rand(4,4)#hidden to hidden
Why = np.random.rand(1,4)#hidden to output
bh = np.random.rand(4)#hidden bias

print(Wxh)
print(Whh)
print(Why)
print(bh)
```

```python
#input contribution
def input_contribution(x, Wxh):
    return np.dot(Wxh, x)

#hidden contribution
def hidden_contribution(h, Whh):
    return np.dot(Whh, h)

#add bias
def add_bias(contribution, b):
    return contribution + b

#apply activation function (tanh)
def apply_activation(x):
    return np.tanh(x)
```

```python
#forward pass for one time step
for t in range(3):
    if t == 0:
        x = x1
        h_prev = h0
    elif t == 1:
        x = X2
        h_prev = h1
    else:
        x = x3
        h_prev = h2

    input_contr = input_contribution(x, Wxh)
    hidden_contr = hidden_contribution(h_prev, Whh)
    total_contr = add_bias(input_contr + hidden_contr, bh)
    h_new = apply_activation(total_contr)

    print(f"Hidden state at time step {t+1}: {h_new}")

    if t == 0:
        h1 = h_new
    elif t == 1:
        h2 = h_new
```
