```python
#hidden size of 4
h0 = np.random.rand(4)

#initialize weights 
Wxh = np.random.rand(4,5)#input to hidden
Whh = np.random.rand(4,4)#hidden to hidden
Why = np.random.rand(1,4)#hidden to output
bh = np.random.rand(4)#hidden bias

print(Wxh)
print(Whh)
print(Why)
print(bh)
```

```python
#input contribution
def input_contribution(x, Wxh):
    return np.dot(Wxh, x)

#hidden contribution
def hidden_contribution(h, Whh):
    return np.dot(Whh, h)

#add bias
def add_bias(contribution, b):
    return contribution + b

#apply activation function (tanh)
def apply_activation(x):
    return np.tanh(x)
```

```python
# List to store hidden states and outputs
hidden_states = []
outputs = []

# Forward pass for 3 time steps
for t in range(3):
    # Select input and previous hidden state
    if t == 0:
        x = x1
        h_prev = h0
    elif t == 1:
        x = x2
        h_prev = h1
    else:
        x = x3
        h_prev = h2

    # Compute contributions
    input_contr = input_contribution(x, Wxh)
    hidden_contr = hidden_contribution(h_prev, Whh)
    total_contr = add_bias(input_contr + hidden_contr, bh)
    h_new = apply_activation(total_contr)

    # Store hidden state
    hidden_states.append(h_new)

    # Compute output
    y_t = Why @ h_new  # shape (1,)
    outputs.append(y_t)

    print(f"Hidden state at time step {t+1}: {h_new}")
    print(f"Output at time step {t+1}: {y_t}")

    # Update variables for next step
    if t == 0:
        h1 = h_new
    elif t == 1:
        h2 = h_new
```
